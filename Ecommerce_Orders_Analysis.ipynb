# 🛒 E-commerce Orders Analysis

**Objective:** Analyze customer orders, revenue trends, and product performance for an e-commerce business using SQL for data extraction and Python for analysis and visualization.

**Tools used:**
- **SQL (PostgreSQL):** data querying — filtering, JOINs, GROUP BY, aggregations
- **Python:** Pandas for analysis, Matplotlib for visualization

**Dataset:** 100 orders across 30 customers, 15 products, and 4 categories (Electronics, Furniture, Sports, Home, Accessories), collected Jan–Jul 2026.

**Tables:**
| Table | Description |
|---|---|
| `customers` | customer_id, customer_name, gender, city, signup_date |
| `orders` | order_id, customer_id, order_date, order_status, payment_method |
| `order_items` | order_item_id, order_id, product_id, quantity, unit_price |
| `products` | product_id, product_name, category, price |

---


````python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

plt.rcParams["figure.facecolor"] = "white"
pd.options.display.float_format = "{:,.2f}".format

# Load raw tables (exported from PostgreSQL)
customers = pd.read_csv("customers.csv")
orders = pd.read_csv("orders.csv", parse_dates=["order_date"])
order_items = pd.read_csv("order_items.csv")
products = pd.read_csv("products.csv")

print(f"Customers: {len(customers)} | Orders: {len(orders)} | Order items: {len(order_items)} | Products: {len(products)}")
````

### Building the joined dataset

To answer product- and customer-level questions, we join `order_items` → `orders` → `products` → `customers` into a single flat table, then compute `line_revenue = quantity × unit_price`.

**Equivalent SQL:**
```sql
SELECT oi.*, o.order_date, o.order_status, o.payment_method,
       p.product_name, p.category,
       c.customer_name, c.city, c.gender
FROM order_items oi
JOIN orders o ON oi.order_id = o.order_id
JOIN products p ON oi.product_id = p.product_id
JOIN customers c ON o.customer_id = c.customer_id;
```


````python
df = (
    order_items
    .merge(orders, on="order_id", how="left")
    .merge(products, on="product_id", how="left")
    .merge(customers, on="customer_id", how="left")
)
df["line_revenue"] = df["quantity"] * df["unit_price"]

# Most business questions only make sense for orders that actually completed
delivered = df[df["order_status"] == "Delivered"]

df.head()
````

---
## 1. Total Revenue & Orders

**SQL:**
```sql
SELECT ROUND(SUM(oi.quantity * oi.unit_price), 2) AS total_revenue
FROM order_items oi
JOIN orders o ON oi.order_id = o.order_id
WHERE o.order_status = 'Delivered';

SELECT COUNT(*) AS total_orders FROM orders WHERE order_status = 'Delivered';
```

````python
total_revenue = delivered["line_revenue"].sum()
total_orders = orders["order_status"].eq("Delivered").sum()
avg_order_value = delivered.groupby("order_id")["line_revenue"].sum().mean()

print(f"Total Revenue: ₹{total_revenue:,.2f}")
print(f"Total Delivered Orders: {total_orders}")
print(f"Average Order Value: ₹{avg_order_value:,.2f}")
````

**Insight:** The business generated **₹13.19L** in revenue from **81 delivered orders**, at an average order value of **₹16,288**.

---
## 2. Monthly Revenue Trend

**SQL:**
```sql
SELECT DATE_TRUNC('month', o.order_date) AS month,
       SUM(oi.quantity * oi.unit_price) AS monthly_revenue
FROM orders o
JOIN order_items oi ON o.order_id = oi.order_id
WHERE o.order_status = 'Delivered'
GROUP BY month
ORDER BY month;
```

````python
monthly_revenue = (
    delivered.assign(month=delivered["order_date"].dt.to_period("M").dt.to_timestamp())
    .groupby("month")["line_revenue"].sum()
    .reset_index()
)

fig, ax = plt.subplots(figsize=(10, 4.5))
ax.plot(monthly_revenue["month"], monthly_revenue["line_revenue"],
        marker="o", color="#0F766E", linewidth=2.5, markersize=7)
ax.fill_between(monthly_revenue["month"], monthly_revenue["line_revenue"], color="#0F766E", alpha=0.08)
ax.set_title("Monthly Revenue Trend", fontsize=13, fontweight="bold", loc="left")
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"₹{x/1000:.0f}K"))
ax.spines[["top", "right"]].set_visible(False)
ax.grid(axis="y", color="#E2E8F0")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
````

**Insight:** Revenue peaked in **January** and dipped through **February and April**, before partially recovering in **May**. July (partial data) is the weakest month. This volatility is worth investigating — was it seasonal demand, a marketing pause, or stockouts?

---
## 3. Revenue by Payment Method

**SQL:**
```sql
SELECT payment_method,
       SUM(oi.quantity * oi.unit_price) AS total_revenue,
       COUNT(DISTINCT o.order_id) AS total_orders
FROM orders o
JOIN order_items oi ON o.order_id = oi.order_id
WHERE o.order_status = 'Delivered'
GROUP BY payment_method
ORDER BY total_revenue DESC;
```

````python
payment_summary = (
    delivered.groupby("payment_method")
    .agg(total_revenue=("line_revenue", "sum"), total_orders=("order_id", "nunique"))
    .sort_values("total_revenue", ascending=False)
)

fig, ax = plt.subplots(figsize=(9, 4.5))
ax.barh(payment_summary.index[::-1], payment_summary["total_revenue"][::-1], color="#0F766E")
ax.set_title("Revenue by Payment Method", fontsize=13, fontweight="bold", loc="left")
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"₹{x/1000:.0f}K"))
ax.spines[["top", "right"]].set_visible(False)
ax.grid(axis="x", color="#E2E8F0")
plt.tight_layout()
plt.show()

payment_summary
````

**Insight:** **Cash on Delivery** is the single largest payment method by revenue, but **Debit Card + Credit Card + UPI combined** actually exceed it — suggesting digital payments dominate in aggregate even though COD leads individually.

## 4. Cash on Delivery Share

**SQL:**
```sql
SELECT ROUND(100.0 * COUNT(*) FILTER (WHERE payment_method = 'Cash on Delivery') / COUNT(*), 2) AS cod_percentage
FROM orders;
```

````python
cod_percentage = round(100 * (orders["payment_method"] == "Cash on Delivery").sum() / len(orders), 2)
print(f"COD Percentage: {cod_percentage}%")
````

## 5. Top 5 Customers by Spend

**SQL:**
```sql
SELECT c.customer_name, SUM(oi.quantity * oi.unit_price) AS total_spent
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id
JOIN order_items oi ON o.order_id = oi.order_id
WHERE o.order_status = 'Delivered'
GROUP BY c.customer_name
ORDER BY total_spent DESC
LIMIT 5;
```

````python
top_customers = (
    delivered.groupby("customer_name")["line_revenue"].sum()
    .sort_values(ascending=False).head(5)
)

fig, ax = plt.subplots(figsize=(9, 4))
ax.barh(top_customers.index[::-1], top_customers.values[::-1], color="#D97706")
ax.set_title("Top 5 Customers by Spend", fontsize=13, fontweight="bold", loc="left")
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"₹{x/1000:.0f}K"))
ax.spines[["top", "right"]].set_visible(False)
ax.grid(axis="x", color="#E2E8F0")
plt.tight_layout()
plt.show()

top_customers
````

````python
top2_share = top_customers.iloc[:2].sum() / total_revenue * 100
print(f"Top 2 customers account for {top2_share:.1f}% of total revenue")
````

**Insight:** The top 2 customers (Manoj, Meera) alone account for a notable share of total revenue — a customer concentration risk worth flagging for retention strategy.

## 6. Best-Selling Products & Category Mix

**SQL:**
```sql
SELECT p.product_name, SUM(oi.quantity * oi.unit_price) AS product_revenue
FROM order_items oi
JOIN orders o ON oi.order_id = o.order_id
JOIN products p ON oi.product_id = p.product_id
WHERE o.order_status = 'Delivered'
GROUP BY p.product_name
ORDER BY product_revenue DESC
LIMIT 8;

SELECT p.category, SUM(oi.quantity * oi.unit_price) AS category_revenue
FROM order_items oi
JOIN orders o ON oi.order_id = o.order_id
JOIN products p ON oi.product_id = p.product_id
WHERE o.order_status = 'Delivered'
GROUP BY p.category
ORDER BY category_revenue DESC;
```

````python
top_products = delivered.groupby("product_name")["line_revenue"].sum().sort_values(ascending=False).head(8)
category_revenue = delivered.groupby("category")["line_revenue"].sum().sort_values(ascending=False)

fig, axes = plt.subplots(1, 2, figsize=(13, 4.5))

axes[0].barh(top_products.index[::-1], top_products.values[::-1], color="#0F766E")
axes[0].set_title("Top Products by Revenue", fontsize=12, fontweight="bold", loc="left")
axes[0].xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"₹{x/1000:.0f}K"))
axes[0].spines[["top", "right"]].set_visible(False)
axes[0].grid(axis="x", color="#E2E8F0")

axes[1].pie(category_revenue.values, labels=category_revenue.index, autopct="%1.0f%%",
            colors=["#0F766E", "#D97706", "#7C3AED", "#0EA5E9", "#94A3B8"],
            wedgeprops={"width": 0.4})
axes[1].set_title("Revenue by Category", fontsize=12, fontweight="bold", loc="left")

plt.tight_layout()
plt.show()
````

**Insight:** The **Laptop** is the single biggest revenue driver — more than 3x the next product. **Electronics** dominates category mix at over half of total revenue, meaning the business is heavily reliant on one category.

## 7. Order Status & Cancellation Rate

**SQL:**
```sql
SELECT order_status, COUNT(*) AS num_orders
FROM orders
GROUP BY order_status;
```

````python
status_counts = orders["order_status"].value_counts()
cancellation_rate = round(100 * status_counts.get("Cancelled", 0) / len(orders), 1)

fig, ax = plt.subplots(figsize=(6, 4.5))
colors = ["#0F766E" if s == "Delivered" else "#DC2626" for s in status_counts.index]
bars = ax.bar(status_counts.index, status_counts.values, color=colors, width=0.5)
ax.set_title("Order Status Breakdown", fontsize=13, fontweight="bold", loc="left")
ax.spines[["top", "right"]].set_visible(False)
ax.grid(axis="y", color="#E2E8F0")
for b in bars:
    ax.text(b.get_x() + b.get_width()/2, b.get_height() + 1, f"{int(b.get_height())}",
            ha="center", fontsize=11, fontweight="bold")
plt.tight_layout()
plt.show()

print(f"Cancellation rate: {cancellation_rate}%")
````

**Insight:** **19% of all orders are cancelled** — nearly 1 in 5. This is a meaningful operational red flag worth investigating further (payment method correlation, product category, or delivery city).

## 8. City-Wise Revenue

**SQL:**
```sql
SELECT c.city, SUM(oi.quantity * oi.unit_price) AS city_revenue
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id
JOIN order_items oi ON o.order_id = oi.order_id
WHERE o.order_status = 'Delivered'
GROUP BY c.city
ORDER BY city_revenue DESC
LIMIT 6;
```

````python
city_revenue = delivered.groupby("city")["line_revenue"].sum().sort_values(ascending=False).head(6)

fig, ax = plt.subplots(figsize=(9, 4.5))
ax.barh(city_revenue.index[::-1], city_revenue.values[::-1], color="#D97706")
ax.set_title("Top Cities by Revenue", fontsize=13, fontweight="bold", loc="left")
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"₹{x/1000:.0f}K"))
ax.spines[["top", "right"]].set_visible(False)
ax.grid(axis="x", color="#E2E8F0")
plt.tight_layout()
plt.show()

city_revenue
````

**Insight:** **Delhi** leads in total revenue, but **Kolkata** generates nearly as much from fewer customers — implying a higher revenue-per-customer in Kolkata worth exploring for targeted marketing.

## 9. Customer Loyalty: Repeat vs One-Time Buyers

**SQL:**
```sql
SELECT customer_id, COUNT(*) AS num_orders
FROM orders
WHERE order_status = 'Delivered'
GROUP BY customer_id;
```

````python
orders_per_customer = delivered.groupby("customer_id")["order_id"].nunique()
repeat_customers = (orders_per_customer > 1).sum()
one_time_customers = (orders_per_customer == 1).sum()

fig, ax = plt.subplots(figsize=(7, 2.2))
ax.barh(["Loyalty"], [one_time_customers], color="#94A3B8", label="One-time")
ax.barh(["Loyalty"], [repeat_customers], left=[one_time_customers], color="#0F766E", label="Repeat")
ax.set_xlim(0, one_time_customers + repeat_customers)
ax.set_yticks([])
ax.legend(loc="upper center", bbox_to_anchor=(0.5, 1.6), ncol=2, frameon=False)
for spine in ax.spines.values():
    spine.set_visible(False)
plt.tight_layout()
plt.show()

print(f"Repeat customers: {repeat_customers}")
print(f"One-time customers: {one_time_customers}")
````

**Insight:** **22 of 28 customers (79%) are repeat buyers.** This is a strong loyalty signal — the business retains customers well once acquired, which is often harder than the initial sale.

## Key Takeaways

1. **Revenue is volatile month-to-month** — no clear consistent growth trend across Jan–Jul 2026, worth investigating for seasonality or operational gaps.
2. **19% cancellation rate** is high and should be root-caused (payment method, product category, or fulfillment delays).
3. **Electronics (57%) and the Laptop product alone drive the majority of revenue** — a concentration risk if that category dips.
4. **Customer loyalty is strong (79% repeat rate)**, but revenue is also concentrated in a few high-value customers — worth balancing acquisition with retention of mid-tier customers.
5. **Kolkata shows the highest revenue-per-customer** despite fewer total customers than Delhi — a potential expansion opportunity.
